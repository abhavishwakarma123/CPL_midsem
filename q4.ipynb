```python
import import_ipynb      #for importing other files in jupyter
import mylibrary_true as ml   #for accessing library
import math              #for math functions exp and sin
from matplotlib import pyplot as plt  #for plotting to compare the convergence
```

```python
#writing the expression

exp = "(4*(math.exp(-1 * x))*math.sin(x)) - 1"
epsilon = 0.0001 #desired accuracy


#trial interval
a = 0.0
b = 1


#solving the root using bisection and regula falsi #Abha Vishwakarma 1911003
root_bis = ml.Roots.bisection(exp, a, b, epsilon, True, True)
root_reg = ml.Roots.regula_falsi(exp, a, b, epsilon, True, True)

print()
print()
print("The root of the equation calculated from bisection = ", round(root_bis, 6))
print("The root of the equation from regula falsi = ", round(root_reg, 6))


#checking if the roots are correct
func = lambda x: eval(exp)

print()
print()
print("Function at root from bisection = ", round(func(root_bis), 6))
print("Function at root from regula falsi = ", round(func(root_reg), 6))

'''We see that regula falsi uses 3 steps less than bisection to find the root to the desired accuracy, hence regula falsi is more efficiet than bisection.'''
```
